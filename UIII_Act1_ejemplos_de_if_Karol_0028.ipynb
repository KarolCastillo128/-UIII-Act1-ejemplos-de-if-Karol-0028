#### Ejemplos de if

```python
# 1. Ejemplo de if simple (Uso de Operador Relacional)
# Este tipo de estructura ejecuta un bloque solo si la condición se cumple

edad = 20 # 1. Define una variable entera llamada edad.
if edad >= 18: # 2. Evalúa si edad es mayor o igual a 18 (operador relacional).
    print("Eres mayor de edad.") # 3. Imprime el mensaje si la condición es verdadera.
```

```python
# 2. Ejemplo de if else (Uso de Operadores Aritméticos y Relacionales)
# La estructura else define un camino alternativo si la condición inicial es falsa

numero = 7 # 1. Asigna el valor 7 a la variable numero.
if numero % 2 == 0: # 2. Calcula el resto de dividir por 2 (%) y verifica si es igual (==) a 0.
    print("El número es par.") # 3. Acción si el resto es cero (número par).
else: # 4. Si la condición del paso 2 es falsa, ejecuta lo siguiente.
    print("El número es impar.") # 5. Acción si el número no es par.
```

```python
# 3. Ejemplo de if elif else (Uso de Operadores Lógicos y Relacionales)
# Se utiliza para manejar múltiples condiciones alternativas de forma jerárquica

cantidad = 8 # 1. Define la variable cantidad con valor 8.
if cantidad >= 10: # 2. Comprueba si la cantidad es 10 o más.
    descuento = 0.2 # 3. Asigna un 20% de descuento si se cumple.
elif cantidad >= 5: # 4. Si falló el paso 2, comprueba si es al menos 5.
    descuento = 0.1 # 5. Asigna un 10% de descuento si esta condición es verdadera.
else: # 6. Si ninguna de las condiciones anteriores fue verdadera.
    descuento = 0 # 7. Define que no hay descuento.
print(f"Tu descuento es {descuento * 100}%") # 8. Muestra el resultado final calculado.
```

```python
# 4. Ejemplo complejo (Uso Combinado de Aritmética, Lógica y Relación)
# Identificación de si un año es bisiesto, lo cual requiere lógica combinada

año = 2024 # 1. Define el año a evaluar.
# 2. Verifica si el año es divisible por 4 y no por 100, O si es divisible por 400.
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(f"{año} es un año bisiesto.") # 3. Mensaje si la lógica booleana resulta True.
else: # 4. En cualquier otro caso.
    print(f"{año} no es un año bisiesto.") # 5. Mensaje si la lógica resulta False.
```

```python
# 5. Ejemplo de Validación de Rangos (Uso de Relación Encadenada)
# Python permite verificar si un número está entre otros dos de forma simplificada

puntuacion = 15 # 1. Define una puntuación de ejemplo.
if 10 <= puntuacion <= 20: # 2. Comprueba si el valor está en el rango de 10 a 20 inclusive.
    print("La puntuación es aceptable.") # 3. Se ejecuta si está dentro del rango.
else: # 4. Si está fuera del rango.
    print("Fuera de rango.") # 5. Se ejecuta si es menor a 10 o mayor a 20.
```
